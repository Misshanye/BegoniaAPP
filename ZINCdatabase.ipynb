# Processing after download

```python
#@title Write a bash for processing
com_file = open('1.sh','w')
com_file.write('''
#!/bin/bash

# 进入每个子文件夹
for dir in */; do
  cd "$dir"
  # 进入内部文件夹
  for inner_dir in */; do
    cd "$inner_dir"
    # 解压所有.gz文件
    for file in *.gz; do
      gunzip "$file"
    done
    # 复制所有解压后的文件到上层文件夹
    mv *.sdf ../../
    cd ../
  done
  cd ../
done

# 合并所有文件
cat *.sdf > all_zinc.sdf
''')
com_file.close()
```

```python
#@title Write a bash for processing
com_file = open('1.sh','w')
com_file.write('''
#!/bin/bash

# 创建 ZKY 文件夹（如果不存在）
mkdir -p ZKY

# 进入每个子文件夹
for dir in */; do
  cd "$dir"
  # 进入内部文件夹
  for inner_dir in */; do
    cd "$inner_dir"
    # 解压所有.gz文件
    for file in *.gz; do
      gunzip "$file"
    done
    # 复制所有解压后的文件到 ZKY 文件夹
    mv *.sdf ../../ZKY/
    cd ../
  done
  cd ../
done

# 进入 ZKY 文件夹并合并所有文件
cd ZKY
cat *.sdf > all_zinc.sdf
''')
com_file.close()
```

```python
import os
import glob

# 指定文件夹路径
folder_path = '/content/ZKY'
output_file = '/content/all_zinc.sdf'

# 列出所有的.sdf文件
sdf_files = glob.glob(os.path.join(folder_path, '*.sdf'))

# 打开输出文件准备写入
with open(output_file, 'w') as outfile:
    for sdf_file in sdf_files:
        # 逐个读取.sdf文件并写入到输出文件
        with open(sdf_file, 'r') as infile:
            content = infile.read()
            outfile.write(content)
            # 如果需要文件之间有分隔，可以加入分隔符
            # outfile.write('\n$$$$$\n')  # 如果每个分子之间有分隔符
```
